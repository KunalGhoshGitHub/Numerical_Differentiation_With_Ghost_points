```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Original_Function = pd.read_csv("Original_Function_File.txt",delimiter=" ",header=None).to_numpy()
```

```python
Grid_Location = pd.read_csv("Grid_Location.txt",delimiter=" ",header=None).to_numpy()
```

```python
Second_Derivative = pd.read_csv("Output_File.txt",delimiter=" ",header=None).to_numpy()
```

```python
plt.plot(Grid_Location,Original_Function)
plt.plot(Grid_Location,Second_Derivative)
plt.legend(["f(x)", "f''(x)"])
plt.xlabel("x")
plt.ylabel("f(x) , f''(x)")
plt.grid()
plt.show()
```
